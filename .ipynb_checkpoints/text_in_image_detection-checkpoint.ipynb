```python
import cv2
import easyocr
import matplotlib.pyplot as plt
import os
img_path='C:/Users/naray/OneDrive/Desktop/AiML/1.png'
img = cv2.imread(img_path)
text_reader=easyocr.Reader(['en'],gpu=False)
text_data=text_reader.readtext(img)
for t in text_data:
    gbox,_text,val=t
    print(gbox)
    img = cv2.rectangle(img,(int(gbox[0][0]),int(gbox[0][1])),(int(gbox[2][0]),int(gbox[2][1])),(0,255,0),thickness=5)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()
```
